```python
# 导入所需的库
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 定义一个函数，用来将Excel文件转换为CSV文件
def convert_excel_to_csv(excel_file, csv_file):
    # 读取Excel文件中的数据
    df = pd.read_excel(excel_file)

    # 创建一个空的DataFrame，用来存储转换后的数据
    df_new = pd.DataFrame(columns=["text", "label"])

    # 遍历每一行和每一列，将非空单元格作为文本和类别
    for i, row in df.iterrows():
        for col in df.columns:
            if row[col]:
                df_new = df_new.append({"text": row[col], "label": col}, ignore_index=True)

    # 保存转换后的数据为CSV文件
    df_new.to_csv(csv_file, index=False)

# 调用函数，将"data.xlsx"转换为"data.csv"
convert_excel_to_csv("/content/merged (1).xlsx", "data.csv")

# 读取CSV文件中的数据
df = pd.read_csv("data.csv")

# 检查并删除空值
df = df.dropna(subset=["text"])

# 将非字符串值转换为字符串
df["text"] = df["text"].astype(str)

# 将文本转换为词袋特征向量
vectorizer = CountVectorizer(binary=True)
X = vectorizer.fit_transform(df["text"]).toarray()

# 将类别标签转换为数值编码
encoder = LabelEncoder()
y = encoder.fit_transform(df["label"])

# 创建并训练决策树分类器
clf = DecisionTreeClassifier()
clf.fit(X, y)

# 给定一个新的句子，预测其类别概率
new_sentence = "第五条 街道办事处设主任一人，按照工作的繁简和管辖区域的大小，设干事若干人，在必要的时候，可以设副主任一广告主、广告经营者、广告发布者从事广告活动，应当遵守法律、法规，诚实信用，公平竞争。"
X_new = vectorizer.transform([new_sentence]).toarray()
y_pred = clf.predict_proba(X_new)

# 打印结果
print("The probabilities of the categories are:")
for label, prob in zip(encoder.classes_, y_pred[0]):
    print(f"{label}: {prob*100:.2f}%")
```
